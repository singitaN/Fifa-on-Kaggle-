#  clean data
## lets clean the dataframe by removing rows with Nan values

```python
'''lets use drop() to remove all the irrelevant columns'''

fifap = fifa.drop(columns=pos)
fifap.info()
```

### lets remove columns that are not useful and alovate it to 'cfifa' (c=clean.) for a clean dataframe

```python
cfifa=fifap.drop(columns="Unnamed: 0")
cfifa.head()
```

```python
cfifa['Value'].head()
```

```python
def singi(string):
    if string[-1] == 'K':
        return float(string[1:-1])*1000
    elif string[-1] == 'M':
        return float(string[1:-1])*1000000
```

```python
cfifa['Value']= [singi(value) for value in cfifa['Value']]
cfifa['Wage'] = [singi(value) for value in cfifa['Wage']]
```
